```python
import yfinance as yf

# Use the Class A shares (voting) ticker
ticker = yf.Ticker("GOOGL")

# 1. Price History (last 30 days)
price_data = ticker.history(period="1mo", interval="1d")

# 2. Company Info and Valuation
info = ticker.info
pe_ratio = info.get("trailingPE")
forward_pe = info.get("forwardPE")
market_cap = info.get("marketCap")
sector = info.get("sector")
industry = info.get("industry")

# 3. Financials
income_statement = ticker.financials
balance_sheet = ticker.balance_sheet
cash_flow = ticker.cashflow

# 4. Earnings
earnings_annual = ticker.earnings
earnings_quarterly = ticker.quarterly_earnings

# 5. Dividends and Splits
dividends = ticker.dividends
splits = ticker.splits

# 6. Fast Info (Quick snapshot)
fast_info = ticker.fast_info

# Display summary
print("=== Google (Alphabet Inc.) Snapshot ===")
print(f"Current Price: {fast_info.get('lastPrice')}")
print(f"Market Cap: {market_cap}")
print(f"P/E Ratio (Trailing): {pe_ratio}")
print(f"P/E Ratio (Forward): {forward_pe}")
print(f"Sector: {sector}")
print(f"Industry: {industry}")
print("\n--- Last 5 Days of Price Data ---")
print(price_data.tail())
print("\n--- Annual Earnings ---")
if earnings_annual is not None and not earnings_annual.empty:
    print(earnings_annual.tail())
else:
    print("No annual earnings data available.")

print("\n--- Latest Income Statement ---")
if income_statement is not None and not income_statement.empty:
    print(income_statement.iloc[:, 0])
else:
    print("No income statement data available.")
```
